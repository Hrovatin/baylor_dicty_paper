## Load gene sets

## Enrichment reports
The enrichment reports show enriched gene sets filtered based on FDR<=0.25 and gene set-gene group overlap>=2.
Barplots show fold enrichment coloured by FDR. Barplot colourscale is consistent across enrichment reports, but barplot x-axis range differs. 

```python
# Gene set enrichment filtering parameters
PADJ=0.25
MIN_OVERLAP=2
# Plotting parameters
MAX_FE_BAR=None
MIN_FDR_BAR=10**(-10)
LFDR_BASE_BAR=10
CMAP_FDR_BAR=['#bede57','#5fb526','#36801d','#015713','#000803']
```

```python
# Legend for fold change barplot
%matplotlib agg
fig,ax=enr.plot_legend_enrichment_bar(cmap=CMAP_FDR_BAR, min_FDR=MIN_FDR_BAR,used_padj=PADJ,base=LFDR_BASE_BAR)
display(fig)
```

### Milestones

```python
data_deseq=pd.read_table(path_milestones_deseq+'AX4/combined.tsv',index_col=0)
data_impulse=pd.read_table(path_milestones_impulse+'DEacrossStages_summary_mainstage_AX4_0.001.tsv',index_col=0)
data=pd.concat([data_deseq,data_impulse],sort=True,axis=1)
```

```python
comparisons=[comparison.replace('_FDR_overall','') for comparison in data.columns[
    data.columns.str.contains('_FDR_overall')]]
max_fdr_de=0.01
min_abs_lfc_de=2
```

```python
%matplotlib agg
figs=[]
for comparison in comparisons:
    for lFC_filter in ['>= ','<= -']:
        query_names=list(data.query( comparison + '_log2FoldChange ' +lFC_filter+ str(min_abs_lfc_de) +
                          ' & ' + comparison + '_FDR_overall <= ' + str(max_fdr_de) +
                          ' & ' + comparison + ' == 1').index)
        comparison_title=comparison+' lFC '+lFC_filter+ str(
            min_abs_lfc_de)+' and FDR_overall <= ' + str(max_fdr_de)
        if(len(query_names)>0):
            enriched=enr.group_diff_enrichment(query_names=query_names,
                                               group=comparison_title,
                                               name_eid=NAME_EID,all_gene_names_eid=ALL_GENE_NAMES_EID, 
                                               gene_sets_ontology=GENE_SETS_ONTOLOGY,
                                               use_annotated_genes=True,
                                               min_overlap=MIN_OVERLAP,padj=PADJ,
                                               make_enrichment_bar=True,
                                               max_FE_bar=MAX_FE_BAR, min_FDR_bar=MIN_FDR_BAR,cmap_FDR_bar=CMAP_FDR_BAR,
                                               lFDR_base_bar=LFDR_BASE_BAR
                                              )
            figs.append(enriched[1])
            display(enriched[1][0])
        else:
            print('***  '+comparison_title+' selected: 0\n')
plt.close('all')  
```
